```python
# Initialize lists to store URL and content
url_list = []
content_list = []
```

```python
# Iterate over each URL
for url in urls:
    # Send a GET request to the URL
    response = requests.get(url)
    
    # Check if the request was successful
    if response.status_code == 200:
        # Parse the HTML content of the page
        soup = BeautifulSoup(response.text, 'html.parser')
        
        # Find all elements with the class "des_container"
        des_containers = soup.find_all(class_="des_container")
        
        # Extract the content from each "des_container"
        for container in des_containers:
            # Extract the text content inside the container
            content = container.get_text()
            # Append URL and content to respective lists
            url_list.append(url)
            content_list.append(content)
    else:
        print(f'Failed to fetch the page {url}. Status code:', response.status_code)
```

```python
# Create a DataFrame from the lists
df = pd.DataFrame({'URL': url_list, 'Content': content_list})

```
